2. (10б) Створіть pie chart, який би показував співвідношення виживших до загиблих.

Відповідно до цієї діаграми більшість пасажирів корабля(61,6% - майже 2/3) не вижили у цій катастрофі, і лише незначну частину(38,4%- приблизно третя частина) вдалось врятувати.

3. (20б) Створіть bar chart, який би показував співвідношення загиблих до виживших для кожної статі. Для groupby використовуйте атрибут as_index=False.


image.png

```python
grouped = df.groupby(['Sex', 'Survived'], as_index=False).size()
grouped.columns = ['Sex', 'Survived', 'Count']
grouped = grouped.sort_values(by=['Sex', 'Survived'])
sexes = grouped['Sex'].unique()
x = np.arange(len(sexes))
width = 0.35
dead = grouped[grouped['Survived'] == 0]['Count'].values
survived = grouped[grouped['Survived'] == 1]['Count'].values
plt.figure(figsize=(8, 6))
plt.bar(x - width/2, dead, width, label='0', color='blue')
plt.bar(x + width/2, survived, width, label='1', color='orange')
plt.xticks(x, sexes)
plt.xlabel('Sex')
plt.ylabel('Count')
plt.title('Survived vs Dead by Sex')
plt.legend(title='Survived')
plt.tight_layout()
plt.show()
```

Ця стовпчикова діаграма відображає розподіл пасажирів корабля Titanic за статтю та статусом виживання. Вона поділена на дві основні категорії: "female" (жінки) та "male" (чоловіки), для кожної з яких побудовано два стовпчики різного кольору — синій (ті, хто не вижив) та помаранчевий (ті, хто вижив). )Загалом на кораблі чоловіків було набагато більше, ніж жінок.
Згідно з цією діаграмою, можна зробити такі висновки, що серед жінок більшість пасажирів вижили. Це видно з того, що помаранчевий стовпчик (виживші) значно вищий за синій. А от серед чоловіків ситуація протилежна: переважна більшість загинули. Синій стовпчик (не вижили) для чоловіків є найвищим на всьому графіку, в той час як помаранчевий (вижили) — найнижчий.
Таким чином, ця діаграма показує, що ймовірність вижити значно залежала від статі пасажира. Жінки мали набагато більший шанс вижити, ніж чоловіки. Це можна пояснити принципом "жінки та діти — першими", який, імовірно, діяв під час евакуації з корабля.

2 — другий клас,
3 — третій клас (найнижчий).
Вісь Y: значення віку пасажирів (Age), від 0 до понад 80 років.
Колір:
Синій колір — пасажири, які не вижили (Survived = 0),

Аналіз отриманих даних:

1-й клас:
Виживші (помаранчева частина) охоплюють велику частину пасажирів віком від 20 до 60 років. У цьому класі більшість дорослих та літніх людей.
Ті, хто не вижив, були також переважно старшими.
Загалом, висока щільність виживання у 1-му класі серед середнього і старшого віку.

2-й клас:
Більшість пасажирів — молоді, обидві групи (виживші та не виживші) мають схожий розподіл, досить багато виживших від 20 до 50 років, але також є багато виживших дітей та підлітків(від 0 до 20 років), а більшість загиблих тільки від 20 до 50.

Більшість пасажирів - молоді.
Виживших небагато, і вони переважно молодші, також є і діти, як і у минулому класі.
Майже повністю відсутні літні люди загалом у 3-ому класі, і серед пасажирів у 3-му класі від 40 до 60 років більшість загинули, а не врятувались.

6. (13б) Побудуйте графік розподілу частот по віку пасажирів.


image.png

```python
ages = df['Age'].dropna()
mean_age = ages.mean()
median_age = ages.median()
plt.figure(figsize=(8, 6))
plt.hist(ages, bins=20, color='skyblue', edgecolor='black')
plt.axvline(median_age, color='red', label='median', linewidth=2)
plt.axvline(mean_age, color='green', label='mean', linewidth=2)
plt.title("Passengers' ages distribution")
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend()
plt.tight_layout()
plt.show()
```

Ця діаграма — гістограма розподілу віку пасажирів корабля Titanic. Вона дозволяє побачити, в яких вікових групах було найбільше пасажирів, а також дає змогу порівняти середнє значення віку (mean) з медіанним (median).

Вісь X: Вік пасажирів (Age) — поділений на інтервали (бінти) від 0 до 80 років.
Вісь Y: Кількість пасажирів у кожному віковому інтервалі (Count).
Стовпчики: Кожен бін (блок) відображає кількість пасажирів певного вікового діапазону.

Вертикальні лінії:
Червона лінія — медіана (значення, що ділить вибірку навпіл).
Зелена лінія — середнє значення (арифметичне середнє віку).

Найбільше пасажирів — у віковій групі 20–30 років.

Особливо помітно, що пік кількості (найвищий стовпчик) припадає приблизно на 22–24 роки.

Медіана (червона лінія) трохи менша за середнє (зелена лінія) - приблизно 30 років.

Є кілька пасажирів старшого віку (60–80 років), які "тягнуть" середнє вгору, але не сильно впливають на медіану.

Діти від 0 до 5 років зустрічайться досить часто(їхня кількість 40), а от літніх людей на кораблі було дуже мало

7. (25б) Поставте бізнес-питання до даних. Дайте на нього відповідь за допомогою візуалізацій. Прокоментуйте отриману відповідь.

Чи існує залежність між класом квитка (Pclass) і шансами на виживання пасажира?

```python
grouped = df.groupby(['Pclass', 'Survived'], as_index=False).size()
grouped.columns = ['Pclass', 'Survived', 'Count']
dead = grouped[grouped['Survived'] == 0]['Count'].values
survived = grouped[grouped['Survived'] == 1]['Count'].values
classes = grouped['Pclass'].unique()
x = range(len(classes))
plt.figure(figsize=(8, 6))
plt.bar(x, dead, label='Did not survive', color='tomato')
plt.bar(x, survived, bottom=dead, label='Survived', color='mediumseagreen')
plt.xticks(x, classes)
plt.xlabel('Ticket Class (Pclass)')
plt.ylabel('Passenger Count')
plt.title('Survival by Passenger Class')
plt.legend()
plt.tight_layout()
plt.show()
```

Ця діаграма відображає розподіл пасажирів за класом квитка (Pclass) і їхнім статусом виживання:

Зелений сегмент — ті, хто вижив.
Червоний сегмент — ті, хто не вижив.
Вісь X — клас квитка: 1, 2, 3.
Вісь Y — кількість пасажирів у кожному класі.

Аналізданих на діаграмі:
1-й клас:
Має високу частку виживших: зелена частина майже вдвічі більша за червону. Це свідчить про найвищі шанси на виживання серед усіх класів.

2-й клас:
Загалом пасажирів приблизно стільки, скільки і в 1 класі.Співвідношення виживших і загиблих приблизно рівне.Шанси на виживання — середні.

3-й клас:
Найбільша кількість пасажирів серед усіх класів. Але червона частина набагато більша за зелену — тобто більшість загинули. Це свідчить про найнижчі шанси на виживання(можливо через найбільшу кількість пасажирів тут).
